$
   A=

```python
A = np.array([[5, 1], [3, 4]])
u1 = [1, 0]
u2 = [0, 1]
```

# Formula $(18)$

\begin{equation}
\tag{18}
\operatorname{det}(\mathbf{A})=\prod_{i} \lambda_{i} \quad \lambda_{i}=\operatorname{eig}(\mathbf{A})
\end{equation}

## Visualizing matrix determinant by tranforming unit vectors

```python
print('Unit circle transformed by A:')
plt.figure(figsize=(6, 6))

matrixToPlot(A)

# plt.plot([0, 5, 5, 0, 0], [0, 0, 4, 4, 0], linestyle='--', color='#5B2C6F')
plotFilledVectors(u1, u2, color='#626F2C', alpha=0.7)
plotFilledVectors(A.dot(u1), A.dot(u2), color='#FFD700', alpha=0.3)

plt.text(5, -0.5, r"$A_{11}$")
plt.text(-2, 4, r"$A_{22}$")

plt.xlim(-7, 7)
plt.ylim(-7, 7)
# plt.axis('scaled')
plt.show()
```

$$V_1 = Ae_1, V2 = Ae_2$$
$$area = |V_1 \times V_2|$$
The area of the shape tranformed from unit square is $|V_1 \times V_2| = A_{11}A_{22} - A_{12}A_{21} = det(A)$

## Visualizing matrix determinant by tranforming eigen vectors

```python
eigVecs = np.linalg.eig(A)[1]
eigVecs
```

```python
plt.figure(figsize=(6, 6))

linearTransformation(A)

eigVecs = np.linalg.eig(A)[1]
eigVals = np.linalg.eig(A)[0]
plt.text(*eigVecs[:, 0]*eigVals[0], r"$\lambda_1$")
plt.text(*eigVecs[:, 1]*eigVals[1], r"$\lambda_2$")

plotFilledVectors(eigVecs[:, 0], eigVecs[:, 1], color='#626F2C', alpha=0.7)
plotFilledVectors(A.dot(eigVecs[:, 0]), A.dot(eigVecs[:, 1]), color='#FFD700', alpha=0.3)

plt.xlim(-6, 6)
plt.ylim(-6, 8)
# plt.axis('scaled')
plt.show()
```

<span class=pquote>
the area of the parallelogram after transformation is $\lambda_1 \times \lambda_2$ times of the area of the parallelogram between engen vectors, thus:
\begin{equation}
\operatorname{det}(\mathbf{A})=\prod_{i} \lambda_{i} \quad \lambda_{i}=\operatorname{eig}(\mathbf{A})
\end{equation}
</span>

# Formula $(20)$

\begin{equation}
\tag{20}
\operatorname{det}\left(\mathbf{A}^{T}\right)=\operatorname{det}(\mathbf{A})
\end{equation}

$
   A^T=
  \left[ {\begin{array}{cc}
   5 & 3 \\
   1 & 4 \\
  \end{array} } \right]
$

## Visualizing by tranforming unit vectors

```python
print('Unit circle transformed by A transpose:')
plt.figure(figsize=(6, 6))

matrixToPlot(A.transpose())

# plt.plot([0, 5, 5, 0, 0], [0, 0, 4, 4, 0], linestyle='--', color='#5B2C6F')
plotFilledVectors(u1, u2, color='#626F2C', alpha=0.7)
plotFilledVectors(A.transpose().dot(u1), A.transpose().dot(u2), color='#FFD700', alpha=0.3)

plt.text(5, -0.5, r"$A_{11}$")
plt.text(-2, 4, r"$A_{22}$")

plt.xlim(-7, 7)
plt.ylim(-7, 7)
# plt.axis('scaled')
plt.show()
```

$$V_1 = A^Te_1, V2 = A^Te_2$$
$$area = |V_1 \times V_2|$$
The area of the shape tranformed from unit square is $|V_1 \times V_2| = A_{11}A_{22} - A_{21}A_{12} = det(A^T)$ \\
Meanwhile, $ det(A) = A_{11}A_{22} - A_{12}A_{21}$, so
\begin{equation}
det(A^T) = det(A)
\end{equation}

## Visualizing by tranforming eigen vectors

```python
plt.figure(figsize=(6, 6))

linearTransformation(A.transpose())

eigVecs = np.linalg.eig(A.transpose())[1]
eigVals = np.linalg.eig(A.transpose())[0]
plt.text(*eigVecs[:, 0]*eigVals[0], r"$\lambda_1$")
plt.text(*eigVecs[:, 1]*eigVals[1], r"$\lambda_2$")

plotFilledVectors(eigVecs[:, 0], eigVecs[:, 1], color='#626F2C', alpha=0.7)
plotFilledVectors(A.transpose().dot(eigVecs[:, 0]), A.transpose().dot(eigVecs[:, 1]), color='#FFD700', alpha=0.3)

plt.xlim(-6, 6)
plt.ylim(-6, 8)
# plt.axis('scaled')
plt.show()
```

<span class=pquote>
the area of the parallelogram after transformation is $\lambda_1 \times \lambda_2$ times of the area of the parallelogram between engen vectors, thus:
\begin{equation}
\operatorname{det}(\mathbf{A^T})=\prod_{i} \lambda_{i} \quad \lambda_{i}=\operatorname{eig}(\mathbf{A^T})
\end{equation}
</span>

<span class=pquote>
\begin{equation}
\operatorname{eig}(\mathbf{A}) \stackrel{?}{=} \operatorname{eig}(\mathbf{A^T})
\end{equation}
</span>

# Formula $(22)$

\begin{equation}
\tag{22}
\operatorname{det}\left(\mathbf{A}^{-1}\right)=1 / \operatorname{det}(\mathbf{A})
\end{equation}

```python
numpy.linalg.inv(A)
```

## Visualizing by tranforming unit vectors

```python
# print('Unit circle transformed by inverse of A:')
plt.figure(figsize=(6, 6))

matrixToPlot(numpy.linalg.inv(A))

# plt.plot([0, 5, 5, 0, 0], [0, 0, 4, 4, 0], linestyle='--', color='#5B2C6F')
plotFilledVectors(u1, u2, color='#626F2C', alpha=0.7)
plotFilledVectors(numpy.linalg.inv(A).dot(u1), numpy.linalg.inv(A).dot(u2), color='#FFD700', alpha=0.3)

plt.xlim(-1, 1)
plt.ylim(-1, 1)
# plt.axis('scaled')
plt.show()
```

## Visualizing by tranforming eigen vectors

```python
plt.figure(figsize=(6, 6))

linearTransformation(numpy.linalg.inv(A))

eigVecs = np.linalg.eig(numpy.linalg.inv(A))[1]
eigVals = np.linalg.eig(numpy.linalg.inv(A))[0]
plt.text(*eigVecs[:, 0]*eigVals[0], r"$\lambda_1$")
plt.text(*eigVecs[:, 1]*eigVals[1], r"$\lambda_2$")

plotFilledVectors(eigVecs[:, 0], eigVecs[:, 1], color='#626F2C', alpha=0.7)
plotFilledVectors(numpy.linalg.inv(A).dot(eigVecs[:, 0]), numpy.linalg.inv(A).dot(eigVecs[:, 1]), color='#FFD700', alpha=0.3)

plt.xlim(-1, 1)
plt.ylim(-1, 1)
# plt.axis('scaled')
plt.show()
```

<span class=pquote>
\begin{equation}
\operatorname{eig}(\mathbf{A^{-1}}) \stackrel{?}{=} \frac{1}{\operatorname{eig}(\mathbf{A^T})}
\end{equation}
</span>
