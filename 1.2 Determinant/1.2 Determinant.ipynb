```python
A = np.array([[5, 1], [3, 4]])
u1 = [1, 0]
u2 = [0, 1]
```

# Formula $(18)$

\begin{equation}
\tag{18}
\operatorname{det}(\mathbf{A})=\prod_{i} \lambda_{i} \quad \lambda_{i}=\operatorname{eig}(\mathbf{A})
\end{equation}

## Visualizing matrix determinant by tranforming unit vectors

```python
print('Unit circle transformed by A:')
plt.figure(figsize=(6, 6))

matrixToPlot(A)

# plt.plot([0, 5, 5, 0, 0], [0, 0, 4, 4, 0], linestyle='--', color='#5B2C6F')
plotFilledVectors(u1, u2, color='#626F2C', alpha=0.7)
plotFilledVectors(A.dot(u1), A.dot(u2), color='#FFD700', alpha=0.3)

plt.text(5, -0.5, r"$A_{11}$")
plt.text(-2, 4, r"$A_{22}$")

plt.xlim(-7, 7)
plt.ylim(-7, 7)
# plt.axis('scaled')
plt.show()
```

$$V_1 = Ae_1, V2 = Ae_2$$
$$area = |V_1 \times V_2|$$
The area of the shape tranformed from unit square is $|V_1 \times V_2| = A_{11}A_{22} - A_{12}A_{21} = det(A)$

# Formula $(20)$

\begin{equation}
\tag{20}
\operatorname{det}\left(\mathbf{A}^{T}\right)=\operatorname{det}(\mathbf{A})
\end{equation}

$
   A^T=
  \left[ {\begin{array}{cc}
   5 & 3 \\
   1 & 4 \\
  \end{array} } \right]
$

## Visualizing by tranforming unit vectors

```python
print('Unit circle transformed by A transpose:')
plt.figure(figsize=(6, 6))

matrixToPlot(A.transpose())

# plt.plot([0, 5, 5, 0, 0], [0, 0, 4, 4, 0], linestyle='--', color='#5B2C6F')
plotFilledVectors(u1, u2, color='#626F2C', alpha=0.7)
plotFilledVectors(A.transpose().dot(u1), A.transpose().dot(u2), color='#FFD700', alpha=0.3)

plt.text(5, -0.5, r"$A_{11}$")
plt.text(-2, 4, r"$A_{22}$")

plt.xlim(-7, 7)
plt.ylim(-7, 7)
# plt.axis('scaled')
plt.show()
```

$$V_1 = A^Te_1, V2 = A^Te_2$$
$$area = |V_1 \times V_2|$$
The area of the shape tranformed from unit square is $|V_1 \times V_2| = A_{11}A_{22} - A_{21}A_{12} = det(A^T)$ 
<br />
Meanwhile, $ det(A) = A_{11}A_{22} - A_{12}A_{21}$, so
\begin{equation}
det(A^T) = det(A)
\end{equation}

## Visualizing by tranforming eigen vectors

```python
plt.figure(figsize=(6, 6))

linearTransformation(A.transpose())

eigVecs = np.linalg.eig(A.transpose())[1]
eigVals = np.linalg.eig(A.transpose())[0]
plt.text(*eigVecs[:, 0]*eigVals[0], r"$\lambda_1$")
plt.text(*eigVecs[:, 1]*eigVals[1], r"$\lambda_2$")

plotFilledVectors(eigVecs[:, 0], eigVecs[:, 1], color='#626F2C', alpha=0.7)
plotFilledVectors(A.transpose().dot(eigVecs[:, 0]), A.transpose().dot(eigVecs[:, 1]), color='#FFD700', alpha=0.3)

plt.xlim(-6, 6)
plt.ylim(-6, 8)
# plt.axis('scaled')
plt.show()
```

<span class=pquote>
the area of the parallelogram after transformation is $\lambda_1 \times \lambda_2$ times of the area of the parallelogram between engen vectors, thus:
\begin{equation}
\operatorname{det}(\mathbf{A^T})=\prod_{i} \lambda_{i} \quad \lambda_{i}=\operatorname{eig}(\mathbf{A^T})
\end{equation}
</span>

<span class=pquote>
\begin{equation}
\operatorname{eig}(\mathbf{A}) \stackrel{?}{=} \operatorname{eig}(\mathbf{A^T})
\end{equation}
</span>

# Formula $(24)$

\begin{equation}
\tag{24}
\operatorname{det}\left(\mathbf{I}+\mathbf{u v}^{T}\right)=1+\mathbf{u}^{T} \mathbf{v}
\end{equation}

```python
e1=[1, 0]
e2=[0, 1]

u=np.array([[2], [1.5]])
v=np.array([[0.5], [1]])
A = np.identity(2) + u.dot(np.transpose(v))
```

## Visualizing by tranforming unit vectors

```python
print(r'Unit circle transformed by inverse of $\left(\mathbf{I}+\mathbf{u v}^{T}\right)$:')
plt.figure(figsize=(6, 6))

matrixToPlot(A)

# plt.plot([0, 5, 5, 0, 0], [0, 0, 4, 4, 0], linestyle='--', color='#5B2C6F')
plotFilledVectors([1, 0], [0, 1], color='#626F2C', alpha=0.7)
plotFilledVectors(A.dot(e1), A.dot(e2), color='#FFD700', alpha=0.3)

plt.text(2, 0.8, r"$\mathbf{e_1} + \mathbf{u}v_1$")
plt.text(2, 2.5, r"$\mathbf{e_2} + \mathbf{u}v_2$")

plt.xlim(-4, 4)
plt.ylim(-4, 4)
# plt.axis('scaled')
plt.show()
```

$\left(\mathbf{I}+\mathbf{u v}^{T}\right) \mathbf{e_1} = \mathbf{e_1} + \mathbf{u}v_1$
<br/>
$\left(\mathbf{I}+\mathbf{u v}^{T}\right) \mathbf{e_2} = \mathbf{e_2} + \mathbf{u}v_2$
<br/>
So,

\begin{align}
\operatorname{det}\left(\mathbf{I}+\mathbf{u v}^{T}\right) &= \left(\mathbf{I}+\mathbf{u v}^{T}\right) \mathbf{e_1} \times \left(\mathbf{I}+\mathbf{u v}^{T}\right) \mathbf{e_2} \\
&= \left( \mathbf{e_1} + \mathbf{u}v_1\right) \times \left(\mathbf{e_2} + \mathbf{u}v_2\right) \\
&= \begin{vmatrix}
   1 + u_1v_1 & u_1v_2 \\
   u_2v_1 & 1+u_2v_2 \\
  \end{vmatrix} \\
&= (1+u_1v_1+u_2v_2+u_1v_1u_2v_2)-u_1v_2u_2v_1 \\
&= 1 +u_1v_1+u_2v_2\\
&= 1 + \mathbf{u^T}\mathbf{v}
\end{align}

# Formula $(25)$


\begin{equation}
\tag{25}
\begin{array}{l}{\text { For } n=2:} \\ {\qquad \operatorname{det}(\mathbf{I}+\mathbf{A})=1+\operatorname{det}(\mathbf{A})+\operatorname{Tr}(\mathbf{A})}\end{array}
\end{equation}

$\left(\mathbf{I}+[\mathbf{u}, \mathbf{v}]\right) \mathbf{e_1} = \mathbf{e_1} + \mathbf{u}$
<br/>
$\left(\mathbf{I}+[\mathbf{u}, \mathbf{v}]\right) \mathbf{e_2} = \mathbf{e_2} + \mathbf{v}$
<br/>
So,

\begin{align}
\operatorname{det}\left(\mathbf{I}+\mathbf{A}\right) &= \left(\mathbf{e_1} + \mathbf{u}\right) \times \left(\mathbf{e_2} + \mathbf{v}\right) \\
&= \begin{vmatrix}
   1 + u_1 & v_1 \\
   u_2 & 1+v_2 \\
  \end{vmatrix} \\
&= (1+u_1 + v_2+u_1v_2)-u_2v_1 \\
&= 1 +(u_1v_2-u_2v_1)+u_1+v_2\\
&= 1 + \operatorname{det}(\mathbf{A}) + \operatorname{Tr}(\mathbf{A})
\end{align}
